````python
# (-3 ~ 12)*6   "-" pay on time, "+" pay delayed
# create new feature : Repay_Status_sum, Previous_Payment_Prior_sum, Statement_sum
Repay_Status_sum = data[["Jan_Repay_Status","Feb_Repay_Status","Mar_Repay_Status","Apr_Repay_Status","May_Repay_Status","Jun_Repay_Status"]]
data['Repay_Status_sum'] = Repay_Status_sum.apply(lambda x: x.sum(), axis=1)

Previous_Payment_Prior_sum=data[["Previous_Payment_Prior_Jan","Previous_Payment_Prior_Feb","Previous_Payment_Prior_Mar","Previous_Payment_Prior_Apr","Previous_Payment_Prior_May","Previous_Payment_Prior_Jun"]]
data["Previous_Payment_Prior_sum"]=Previous_Payment_Prior_sum.apply(lambda x: x.sum(), axis=1)

Statement_sum=data[["Jan_Statement","Feb_Statement","Mar_Statement","Apr_Statement","May_Statement","Jun_Statement"]]
data["Statement_sum"]=Statement_sum.apply(lambda x: x.sum(), axis=1)
data.head()
````
